# Problem 3:

The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain; Answer the questions below based on the given information; (Provide an appropriate visual representation of your answers, without which marks will be deducted)

3.1 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?

```python
z=(3.17-5)/1.5
z
```

```python
stats.norm.cdf(-1.22)
```

```python
# Parameters for the normal distribution
mean = 5
std_dev = 1.5

# Calculate z-score
z = (3.17 - mean) / std_dev

# Calculate the probability using the cumulative distribution function (CDF)
p_value = -stats.norm.cdf(-z)

# Create an array of x values
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)

# Calculate the corresponding y values from the normal distribution
y = stats.norm.pdf(x, mean, std_dev)

# Plot the normal curve
plt.plot(x, y)
plt.fill_between(x, y, 0, where=(x <= 3.17), color='orange', alpha=0.3)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')

# Mark the point on the curve
plt.annotate(f'z = {z:.2f}', xy=(3.17, stats.norm.pdf(3.17, mean, std_dev)), xytext=(3.17, 0.12),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Print the calculated P-value
plt.text(mean - 3 * std_dev, 0.2, f'P(z < -1.22) = {p_value:.4f}')

# Show the plot
plt.show()
```

```python
z1=(5.5-5)/1.5
z1
```

```python
z2=(5-5)/1.5
z2
```

```python
stats.norm.cdf(0.33)-stats.norm.cdf(0)
```

```python

# Parameters for the normal distribution
mean = 5
std_dev = 1.5

# Calculate z-scores
z1 = (5.5 - mean) / std_dev
z2 = (5 - mean) / std_dev

# Calculate the probabilities using the cumulative distribution function (CDF)
p_value = stats.norm.cdf(z1) - stats.norm.cdf(z2)

# Create an array of x values
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)

# Calculate the corresponding y values from the normal distribution
y = stats.norm.pdf(x, mean, std_dev)

# Plot the normal curve
plt.plot(x, y)
plt.fill_between(x, y, 0, where=((x >= 5) & (x <= 5.5)), color='orange', alpha=0.3)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')

# Mark the points on the curve
plt.annotate(f'z1 = {z1:.2f}', xy=(5.5, stats.norm.pdf(5.5, mean, std_dev)), xytext=(6.5, 0.35),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'z2 = {z2:.2f}', xy=(5, stats.norm.pdf(5, mean, std_dev)), xytext=(4, 0.35),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Print the calculated P-value
plt.text(mean - 3 * std_dev, 0.4, f'P(5 <= z <= 5.5) = {p_value:.4f}')

# Show the plot
plt.show()
```

# Problem 4:

Grades of the final examination in a training course are found to be normally distributed, with a mean of 77 and a standard deviation of 8.5. Based on the given information answer the questions below.

4.1 What is the probability that a randomly chosen student gets a grade below 85 on this exam?



```python
z5=(85-77)/8.5
z5
```

```python
1-stats.norm.cdf(0.94)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for the normal distribution
mean = 77
std_dev = 8.5

# Calculate z-score for z5
z5 = (85 - mean) / std_dev

# Calculate the probability using the cumulative distribution function (CDF)
prob = 1 - stats.norm.cdf(0.94)

# Create an array of x values
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)

# Calculate the corresponding y values from the normal distribution
y = stats.norm.pdf(x, mean, std_dev)

# Plot the normal curve
plt.plot(x, y)
plt.fill_between(x, y, 0, where=(x >= 85), color='orange', alpha=0.3)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')

# Mark the point z5 on the curve
plt.annotate(f'z5 = {z5:.2f}', xy=(85, stats.norm.pdf(85, mean, std_dev)), xytext=(85, 0.02), 
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Print the calculated probability
plt.text(mean - 3 * std_dev, 0.18, f'P(z > 0.94) = {prob:.4f}')

# Show the plot
plt.show()
```

4.3 What should be the passing cut-off so that 75% of the students clear the exam?

```python
stats.norm.ppf(0.25,loc=77,scale=8.5)
```

```python
# Parameters for the normal distribution
mean = 77
std_dev = 8.5

# Calculate the z-score corresponding to the given P-value
p_value = 0.25
z = stats.norm.ppf(p_value, loc=mean, scale=std_dev)

# Create an array of x values
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)

# Calculate the corresponding y values from the normal distribution
y = stats.norm.pdf(x, mean, std_dev)

# Plot the normal curve
plt.plot(x, y)
plt.fill_between(x, y, 0, where=(x <= z), color='orange', alpha=0.3)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')

# Mark the point on the curve
plt.annotate(f'z = {z:.2f}', xy=(z, stats.norm.pdf(z, mean, std_dev)), xytext=(z-3, 0.02),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Print the calculated z-score
plt.text(mean + 1.5 * std_dev, 0.06, f'P(x <= {z:.2f}) = {p_value}')

# Show the plot
plt.show()
```

# Ouestion-6

 Aquarius health club, one of the largest and most popular cross-fit gyms in the country has been advertising a rigorous program for body conditioning. The program is considered successful if the candidate is able to do more than 5 push-ups, as compared to when he/she enrolled in the program. Using the sample data provided can you conclude whether the program is successful? (Consider the level of Significance as 5%) Note that this is a problem of the paired-t-test. Since the claim is that the training will make a difference of more than 5, the null and alternative hypotheses must be formed accordingly.

```python
mydata1 = pd.read_csv('Aquarius_gym.csv')
mydata1.head()
```

```python
# Alpha= 0.05, n=101
# H0<=5
# Ha>5

t_statistic, p_value  = ttest_ind(mydata1['Before'],mydata1['After'])
print('tstat',t_statistic)    
print('P Value',p_value) 
```

Since the p-value is less than the significance level, we reject the null hypothesis. Therefore, we can conclude that the training program is successful in achieving a difference of more than 5 push-ups.
